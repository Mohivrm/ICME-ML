```python
import cdflib
import matplotlib.pyplot as plt

# Load the CDF file

cdf = cdflib.CDF('wind_h1s_swe_20031120000112_20031120235922.cdf')

# Access the data for the 'BX', 'BY', and 'BZ' variables
BX_C = cdf.varget('BX')
BY_C = cdf.varget('BY')
BZ_C = cdf.varget('BZ')

PX_C = cdf.varget('Proton_VX_moment')
PY_C = cdf.varget('Proton_VY_moment')
PZ_C = cdf.varget('Proton_VZ_moment')


Np_C = cdf.varget('Proton_Np_moment')

BXYZ = BX_C + BY_C + BZ_C


# Calculate the magnitude of the magnetic field
Bmag = (BX_C**2 + BY_C**2 + BZ_C**2)**0.5

# Calculate the magnitude of the Proton moment field
Vp = (PX_C**2 + PY_C**2 + PZ_C**2)**0.5

epoch = cdf.varget('Epoch')
trange = cdflib.cdfepoch.to_datetime(epoch)


# Rotate x-axis labels for better readability
fig, axs = plt.subplots(4, 1, figsize=(20, 15))

axs[0].plot(trange, Bmag)
axs[0].set_ylabel('$B_{\mathrm{mag}}$ (nT)')


axs[1].plot(trange, Np_C)
axs[1].set_ylabel('Proton Np Moment')


# Plot Bvec

# Plot bx, by, bz separately
axs[2].plot(trange, BX_C, label='$B_x$ (nT)', color='red')
axs[2].plot(trange, BY_C, label='$B_y$ (nT)', color='blue')
axs[2].plot(trange, BZ_C, label='$B_z$ (nT)', color='green')

axs[2].set_ylabel('$B_{\mathrm{vec}}$ and Components (nT)')

axs[2].legend()


axs[3].plot(trange, Vp, color='magenta')


plt.xlabel('Time')
plt.xticks(rotation=20)  
# Rotate x-axis labels by 20 degrees
plt.tight_layout()
plt.show()
```
